Because of issues of kernel crashes due to the volume of data, take a sample of 200000 instances for processing.

There are a lot of missing values in the data. Because we have a lot of data, it is okay to drop variables that have 60% or more missing values. I proceed to make a list of such variables and drop them from the analysis.

```
# Make a list of variables with 60% or more missing values

drop_missing_values = []
for column_name in cat_sample.columns:
    numerator = cat_sample[column_name].isnull().sum()
    denominator = numerator + cat_sample[column_name].count()
    if (numerator*100/denominator) >= 60:
        drop_missing_values.append(column_name)
```

```
# Drop columns with 60% or more of missing values

cat_sample.drop(drop_missing_values, axis = 1, inplace = True)
```

Information value measures how much information an independent variable provides to explaining a dependent variable. Information value requires weight of evidence for its calculation. Weight of evidence itself is another metric that quantifies the predictive ability of an independent variable in relation to the dependent variable. The information value is a weighted average of weight of evidence. Information value ranges between 0 and 1. To be considered a good predictor of a dependent variable, the information value of an independent variable must be at least greater or equal to 0.2 (ie. IV >= 0.2, at least 20% worth of information value).

These computations are not difficult but they are tedious because a series of calculations are needed each time to compute weight of evidence and information value. To make this easier, a function is defined below that calculates weight of evidence (WoE) and Information Value (IV) for all the categorical variables and those variables that meet the decision condition (ie. IV >= 0.2) will be seleceted.

To calucualte information value, I need the target variable, hence, I declare the target and split up the data.

```
# Declare and insert the target variable into the dataset

cat_sample['good_bad'] = np.where(cat_sample['loan_status'].isin(['Charged Off', 'Default', 
                                                                   'Late (16-30 days)',
                                                                   'Does not meet the credit policy. Status:Charged Off',
                                                                   ]), 0, 1)
```

```
# Split up the data into train and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cat_sample.drop('good_bad', axis = 1), cat_sample['good_bad'])
```

```
# Name the train sets so they can be used in the calculation of weight of evidence and information value

df_prepr = X_train
df_target = y_train
```

```
# Quick view dataframe

df_prepr.head()
```

```
# Define a function to calculate weight of evidence and information value

def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```
